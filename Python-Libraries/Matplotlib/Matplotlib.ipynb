## ðŸ”¹ 1. Import Matplotlib
```python
import matplotlib.pyplot as plt

````python
import matplotlib.pyplot as plt
````

##### ðŸ”¹ 2. Basic Line Plot

````python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Simple Line Plot")
plt.show()
````

##### ðŸ”¹ 3. Bar Chart

````python
names = ["Ali", "Ahmed", "Sara"]
marks = [85, 90, 88]

plt.bar(names, marks)
plt.title("Marks Bar Chart")
plt.show()
````

##### ðŸ”¹ 4. Histogram

````python
data = [10, 20, 20, 30, 40, 40, 40]

plt.hist(data, bins=5)
plt.title("Histogram")
plt.show()
````

##### ðŸ”¹ 5. Scatter Plot

````python
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
````

##### ðŸ”¹ 6. Multiple Lines

````python
plt.plot(x, y, label="Line 1")
plt.plot(x, [i*2 for i in y], label="Line 2")
plt.legend()
plt.show()
````

##### ðŸ”¹ 7. Figure Size

````python
plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.show()
````

##### ðŸ”¹ 8. Subplots

````python
plt.subplot(1, 2, 1)
plt.plot(x, y)

plt.subplot(1, 2, 2)
plt.bar(names, marks)

plt.show()
````

##### ðŸ”¹ 9. Save Figure

````python
plt.savefig("plot.png")
````

##### ðŸ”¹ 10. Real-World Example

````python
months = ["Jan", "Feb", "Mar"]
sales = [20000, 25000, 30000]

plt.plot(months, sales, marker='o')
plt.title("Monthly Sales")
plt.show()
````
