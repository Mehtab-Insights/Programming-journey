##### ðŸ”¹ 2. Pandas Series

````python
data = pd.Series([10, 20, 30, 40])
print(data)
````

##### ðŸ”¹ 3. Pandas DataFrame

````python
data = {
    "Name": ["Ali", "Ahmed", "Sara"],
    "Age": [20, 22, 21],
    "Marks": [85, 90, 88]
}

df = pd.DataFrame(data)
print(df)
````

##### ðŸ”¹ 4. Reading Data

````python
df = pd.read_csv("data.csv")
````

##### ðŸ”¹ 5. Data Inspection

````python
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
````

##### ðŸ”¹ 6. Selecting Data

````python
print(df["Name"])
print(df[["Name", "Marks"]])
````

Row selection:

````python
print(df.loc[0])
print(df.iloc[1])
````

##### ðŸ”¹ 7. Filtering Data

````python
print(df[df["Marks"] > 85])
print(df[(df["Age"] > 20) & (df["Marks"] > 80)])
````

##### ðŸ”¹ 9. Handling Missing Values

````python
df.isnull()
df.dropna()
df.fillna(0)
````

##### ðŸ”¹ 10. GroupBy

````python
df.groupby("Grade")["Marks"].mean()
````

##### ðŸ”¹ 11. Sorting

````python
df.sort_values(by="Marks", ascending=False)
````

##### ðŸ”¹ 12. Apply Function

````python
df["Marks"] = df["Marks"].apply(lambda x: x + 5)
````
