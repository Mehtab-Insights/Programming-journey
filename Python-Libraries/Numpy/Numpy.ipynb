# ðŸ“Œ NumPy Complete Guide: Beginner to Advanced

NumPy (Numerical Python) is the core library for numerical computing in Python.  
It provides fast, efficient arrays and mathematical functions.

---

## ðŸ”¹ 1. Import NumPy
```python

````python
import numpy as np
````

### ðŸ”¹ 2. Creating NumPy Arrays

````python
arr1 = np.array([1, 2, 3, 4, 5])
print(arr1)
````

##### 2D Array

````python
arr2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
print(arr2)
````

##### 3D Array

````python
arr3 = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
print(arr3)
````

### ðŸ”¹ 3. Array Properties

````python
print(arr2.ndim)     # Number of dimensions
print(arr2.shape)   # Rows and columns
print(arr2.size)    # Total elements
print(arr2.dtype)   # Data type
````

### ðŸ”¹ 4. Special Arrays

````python
zeros = np.zeros((3, 3))
ones = np.ones((2, 4))
identity = np.eye(3)

print(zeros)
print(ones)
print(identity)
````

### ðŸ”¹ 5. Range & Random Arrays

##### Range

````python
arr_range = np.arange(1, 11, 2)
print(arr_range)
````

##### Linspace

````python
lin = np.linspace(0, 10, 5)
print(lin)
````

##### Random Arrays

````python
rand1 = np.random.rand(3, 3)
rand2 = np.random.randint(1, 100, (3, 3))

print(rand1)
print(rand2)
````

### ðŸ”¹ 6. Array Indexing & Slicing

##### 1D Indexing

````python
arr = np.array([10, 20, 30, 40, 50])

print(arr[0])
print(arr[-1])
print(arr[1:4])
````

##### 2D Indexing

````python
mat = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(mat[0, 1])
print(mat[:, 1])
print(mat[1, :])
````

### ðŸ”¹ 7. Reshaping Arrays

````python
arr = np.arange(1, 13)
reshaped = arr.reshape(3, 4)
print(reshaped)
````

##### Flatten

````python
flat = reshaped.flatten()
print(flat)
````

### ðŸ”¹ 8. Mathematical Operations

````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a ** 2)
````

### ðŸ”¹ 9. Universal Functions (ufuncs)

````python
arr = np.array([1, 4, 9, 16])

print(np.sqrt(arr))
print(np.log(arr))
print(np.exp(arr))
print(np.sin(arr))
````

### ðŸ”¹ 10. Aggregation Functions

````python
data = np.array([10, 20, 30, 40, 50])

print(np.sum(data))
print(np.mean(data))
print(np.max(data))
print(np.min(data))
print(np.std(data))
````

### ðŸ”¹ 11. Boolean Indexing

````python
arr = np.array([10, 25, 30, 5, 40])

print(arr[arr > 20])
print(arr[arr % 2 == 0])
````

### ðŸ”¹ 12. Copy vs View

````python
arr = np.array([1, 2, 3, 4])

view_arr = arr.view()
copy_arr = arr.copy()

arr[0] = 100

print(arr)
print(view_arr)
print(copy_arr)
````

### ðŸ”¹ 13. Stacking & Splitting Arrays

##### Stacking

````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.hstack((a, b)))
print(np.vstack((a, b)))
````

##### Splitting

````python
arr = np.array([1, 2, 3, 4, 5, 6])
print(np.split(arr, 3))
````

### ðŸ”¹ 14. Sorting & Searching

````python
arr = np.array([30, 10, 50, 20])

print(np.sort(arr))
print(np.where(arr == 50))
````

### ðŸ”¹ 15. Linear Algebra (Advanced)

````python
A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [5, 6],
    [7, 8]
])

print(np.dot(A, B))
print(np.linalg.det(A))
print(np.linalg.inv(A))
````

### ðŸ”¹ 16. Broadcasting

````python
arr = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

scalar = 10
print(arr + scalar)
````

### ðŸ”¹ 17. Real-World Example: Student Marks Analysis

````python
marks = np.array([
    [85, 90, 88],
    [70, 75, 80],
    [92, 95, 94],
    [60, 65, 58]
])

print(np.mean(marks, axis=1))  # Student-wise average
print(np.mean(marks, axis=0))  # Subject-wise average
print(np.max(marks))           # Top score
print(marks[marks > 80])       # Scores above 80
````

### ðŸ”¹ 18. Performance: NumPy vs Python List

````python
import time

size = 1_000_000
list_data = list(range(size))
np_data = np.arange(size)

start = time.time()
list_result = [x * 2 for x in list_data]
print("List time:", time.time() - start)

start = time.time()
np_result = np_data * 2
print("NumPy time:", time.time() - start)
````
