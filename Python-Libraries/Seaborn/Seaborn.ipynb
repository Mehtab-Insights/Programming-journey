````python
import seaborn as sns
import matplotlib.pyplot as plt
````

##### ðŸ”¹ 2. Load Sample Dataset

````python
df = sns.load_dataset("tips")
print(df.head())
````

##### ðŸ”¹ 3. Line Plot

````python
sns.lineplot(x="total_bill", y="tip", data=df)
plt.show()
````

##### ðŸ”¹ 4. Bar Plot

````python
sns.barplot(x="day", y="total_bill", data=df)
plt.show()
````

##### ðŸ”¹ 5. Count Plot

````python
sns.countplot(x="day", data=df)
plt.show()
````

##### ðŸ”¹ 6. Histogram

````python
sns.histplot(df["total_bill"], bins=10)
plt.show()
````

##### ðŸ”¹ 7. Box Plot

````python
sns.boxplot(x="day", y="total_bill", data=df)
plt.show()
````

##### ðŸ”¹ 8. Violin Plot

````python
sns.violinplot(x="day", y="total_bill", data=df)
plt.show()
````

##### ðŸ”¹ 9. Scatter Plot

````python
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.show()
````

##### ðŸ”¹ 10. Heatmap

````python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.show()
````

##### ðŸ”¹ 11. Pair Plot

````python
sns.pairplot(df)
plt.show()
````

##### ðŸ”¹ 12. Real-World Example

````python
sns.barplot(x="time", y="tip", hue="sex", data=df)
plt.show()
````
